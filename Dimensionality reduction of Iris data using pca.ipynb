IRIS - A small classic dataset from Fisher, 1936. One of the earliest known datasets used for evaluating classification methods.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
iris
```

```python
# Standardize the features (optional, but recommended for PCA)
from sklearn.preprocessing import StandardScaler
X_standardized = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_standardized)

# Create a DataFrame with the principal components and target variable
df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
df['Target'] = y

# Visualize the data in 2D
plt.figure(figsize=(8, 6))
targets = list(iris.target_names)
colors = ['r', 'g', 'b']

for target, color in zip(targets, colors):
    indices_to_keep = df['Target'] == iris.target_names.tolist().index(target)
    plt.scatter(df.loc[indices_to_keep, 'Principal Component 1'],
                df.loc[indices_to_keep, 'Principal Component 2'],
                c=color, s=50)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(targets)
plt.title('PCA of Iris Dataset')
plt.show()
```
